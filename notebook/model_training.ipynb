```python
## fetching the dataframe
import pandas as pd

df = pd.read_csv("data/cleaned_breast_cancer.csv")

## display the head
df.head()
```

```python
X = df.drop("diagnosis",axis=1)
y = df["diagnosis"]
```

```python
X.select_dtypes(exclude="O").columns.to_list()
```

```python
X.head()
```

```python
## creating a pipeline 
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
num_pipeline = Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="median")),
        ("scaler",StandardScaler()),
        ("PCA",PCA(n_components=0.95))

    ]
)

preprocessor = ColumnTransformer([
    ("num_pipeline",num_pipeline,X.columns)
])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,stratify=y,random_state=42)
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score

def eval_model(true,predict):
    accuracy = accuracy_score(true,predict)
    class_report = classification_report(true,predict)
    confi_matrix = confusion_matrix(true,predict)
    roc_auc = roc_auc_score(true,predict)

    return accuracy,class_report,confi_matrix,roc_auc
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
models = {
    "LogisticRegression":LogisticRegression(),
    "SVC":SVC(),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "RandomForestClassifier":RandomForestClassifier()

}

param_grid = {
    "LogisticRegression":{
        "penalty":["l1","l2","elasticnet"],
        "C":[1.0,1.2,1.4,1.6],
        "solver":["sag","saga"]
    },
    "SVC":{
        "kernel":["rbf","poly","linear","sigmoid"],
        "C":[1.0,1.2,1.4,1.6]

    },
    "DecisionTreeClassifier":{
        "criterion":["gini","entropy"],
        "splitter":["best","random"],
        "max_depth":[None,10,20,30,40],
        "min_samples_split":[2,3,4,5],
        "min_samples_leaf":[1,2,3,4,5]

    },
    "RandomForestClassifier":{
        "n_estimators":[100,200,300,400],
        "criterion":["gini","entropy"],
        "max_depth":[None,10,20,30,40],
        "min_samples_split":[2,3,4,5],
        "min_samples_leaf":[1,2,3,4,5]
    }
}

best_score = -1
best_estimator = None

for model_name,model in models.items():
    print(f"{model_name}_evaluation started...")
    gridsearch = GridSearchCV(estimator=model,param_grid=param_grid[model_name],verbose=True,n_jobs=-1,cv=3)
    gridsearch.fit(X_train,y_train)
    y_pred = gridsearch.predict(X_test)

    best_model_score = gridsearch.best_score_
    best_model_estimator = gridsearch.best_estimator_
    best_model_params = gridsearch.best_params_

    print(f"{model_name}_best score: {best_model_score} and its best param is: {best_model_params}")

    accuracy,class_report,confi_matrix,roc_auc = eval_model(y_test,y_pred)

    print(f"{model_name}_Accuracy \n",accuracy)
    print(f"{model_name}_Classification Report \n",class_report)
    print(f"{model_name}_Confusion Matrix \n",confi_matrix)
    print(f"{model_name}_roc_auc_score \n",roc_auc)

    if best_model_score > best_score:
        best_score = best_model_score
        best_estimator = best_model_estimator

print(f"Best estimator: {best_estimator} and its best score: {best_score}")
```
